# 2.Outlier detection and handling

```python
# Boxplot of the entire dataset
plt.figure(figsize=(25,25))
sns.boxplot(data= data)
plt.xticks(rotation = 90);
```

```python
data.columns
```

```python
def outlier_rem(column):
	Q1=np.percentile(data[column],25,interpolation='midpoint')
	Q2=np.percentile(data[column],50,interpolation='midpoint')
	Q3=np.percentile(data[column],75,interpolation='midpoint')
	IQR = Q3 - Q1
	lowlim=Q1-1.5*IQR
	uplim=Q3+1.5*IQR
	print(lowlim)
	print(uplim)
	outlier = []
	for x in data[column]:
    		if ((x>uplim) or (x,lowlim)):
        		outlier.append(x)
	ind1=data[column]>uplim
	x=data.loc[ind1].index

	data.drop(x,inplace = True)
	return

columns=['sensor_00', 'sensor_01', 'sensor_02', 'sensor_03', 'sensor_04',
       'sensor_05', 'sensor_06', 'sensor_07', 'sensor_08', 'sensor_09',
       'sensor_10', 'sensor_11', 'sensor_12', 'sensor_13', 'sensor_14',
       'sensor_16', 'sensor_17', 'sensor_18', 'sensor_19', 'sensor_20',
       'sensor_21', 'sensor_22', 'sensor_23', 'sensor_24', 'sensor_25',
       'sensor_26', 'sensor_27', 'sensor_28', 'sensor_29', 'sensor_30',
       'sensor_31', 'sensor_32', 'sensor_33', 'sensor_34', 'sensor_35',
       'sensor_36', 'sensor_37', 'sensor_38', 'sensor_39', 'sensor_40',
       'sensor_41', 'sensor_42', 'sensor_43', 'sensor_44', 'sensor_45',
       'sensor_46', 'sensor_47', 'sensor_48', 'sensor_49', 'sensor_50',
       'sensor_51']
for x in columns:
	outlier_rem(x)
```

# scaling on the feature set

```python
x=data.drop(['machine_status'],axis=1)
y=pd.DataFrame(data['machine_status'])
```

```python
x.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
data = scalar.fit_transform(data)
```

```python
# Splitting the data set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

# Modelling

```python
# Model 1 : Random Forest Model

from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import f1_score,confusion_matrix,accuracy_score
print('Confusion matrix : \n',confusion_matrix(y_test,y_pred))
print('F1_score :',round(f1_score(y_test,y_pred,average='micro'),3))
print('Accuracy :',round(accuracy_score(y_test,y_pred),3)) 
```

```python
# Model 2 : Logistic Regression

from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
predictions= logmodel.predict(x_test)
print('Accuracy is: ',accuracy_score(y_test,predictions))
```

```python
# Model 3 : DecisionTreeClassifier

from sklearn.tree  import DecisionTreeClassifier
dtmodel = DecisionTreeClassifier()
dtmodel.fit(x_train,y_train)
predictions= dtmodel.predict(x_test)
print('Accuracy is: ',accuracy_score(y_test,predictions))
```

# LinearRegression

```python
from sklearn import linear_model
lr = linear_model.LinearRegression()

model = lr.fit(x_train,y_train)
predictions= model.predict(x_test)

print(y_test)
print(predictions)
```
